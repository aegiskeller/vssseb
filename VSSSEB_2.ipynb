```python
import pandas as pd
from astroquery.simbad import Simbad
from astropy.table import Table
from astroquery.vizier import Vizier
import re
```

```python
df = pd.DataFrame()
```

```python
df = pd.read_excel('VSS SEB observed targets.xlsx', sheet_name='Data')
```

```python
df.to_csv('original.csv', index=False)
```

```python
df.head()
```

```python
len(df)
```

```python
df['Const', 'VarID'] = df['Star'].str.split()
```

```python
object_names = []
for i in range(len(df)):
    details = df['Const', 'VarID'].iloc[i]
    if (len(details) == 2):
        object_name = details[1] +' '+ details[0]
    if (len(details) > 2):
        if (details[2].lower() == "secondary" or details[2].lower() == "primary"):
            object_name = details[1] +' '+ details[0]
        else:
            object_name = details[1]+details[2]# +' '+ details[0]
    object_names.append(object_name)
```

```python
df['name'] = object_names
```

```python
df['SCRIPT_NUMBER_ID'] = df.index
```

```python
print(object_names)
```

```python
vizier = Vizier()
gcvs_info = Vizier(catalog='B/gcvs/gcvs_cat').query_constraints(GCVS=object_names)[0]
```

```python
gcvs_info
```

```python
simbad_info = Simbad.query_objects(object_names)
```

```python
simbad_df = simbad_info.to_pandas()
```

```python
simbad_df
```

```python
gcvs_df = gcvs_info.to_pandas()
```

```python
gcvs_df.info()
```

```python
def replace_extra_spaces(text):
    return re.sub(r'\s+', ' ', text).strip()

gcvs_df['VarName'] = gcvs_df['VarName'].apply(replace_extra_spaces)
```

```python
part1=pd.merge(df, simbad_df, on='SCRIPT_NUMBER_ID', how='left')
```

```python
result_df = pd.merge(part1, gcvs_df, left_on='name', right_on='VarName', how='left')
```

```python
result_df.info()
```

```python
result_df['CONS'] = result_df[('Const', 'VarID')].apply(lambda x: x[0])
```

<hr>

Now to build the dataframe that contains the data we want 

```python
ephem_df = pd.DataFrame()
```

```python
ephem_df = ephem_df.assign(CONS=result_df['CONS'])
ephem_df = ephem_df.assign(NAME=result_df['Star'])
ephem_df['PS'] = "p"
ephem_df['PS'].loc[result_df['Star'].str.contains('Primary')] = "p"
ephem_df['PS'].loc[result_df['Star'].str.contains('Secondary')] = "s"
```

```python
ephem_df['RATING'] = 10
```

```python
ephem_df = ephem_df.assign(RA2000=result_df['RA'].str.replace(' ', '').str.slice(0, -1))
```

```python
ephem_df = ephem_df.assign(DE2000=result_df['DEC'].str.replace(' ', '').str.slice(0, -2))
```

```python
ephem_df = ephem_df.assign(M0=result_df['Epoch (HJD)']- 2400000.5)
```

```python
ephem_df = ephem_df.assign(PER=result_df['Period (d)'])
```

```python
ephem_df = ephem_df.assign(MAX=result_df['magMax'])
```

```python
ephem_df = ephem_df.assign(PRIM=result_df['magMax'])
```

```python
ephem_df = ephem_df.assign(SEK=result_df['magMax'])
```

```python
ephem_df['BRANCH'] = 'V'
ephem_df['LD'] = ''
ephem_df['SD'] = ''
ephem_df['GSC'] = ''
ephem_df = ephem_df.assign(TYPE=result_df['VarType'])
ephem_df['N10'] = '1'
ephem_df['N'] = '1'
```

```python
ephem_df
```

```python
# Define column formatters
formatters = {
    'CONS': '{:6}'.format,  
    'NAME': '{:<17}'.format,  
    'PS': '{:<3}'.format,
    'RATING': '{:<7}'.format,  
    'RA2000': '{:<9}'.format,  
    'DE2000': '{:<8}'.format,
    'M0': '{:<13}'.format,  
    'PER': '{:<13}'.format,  
    'MAX': '{:<9}'.format,
    'PRIM': '{:<9}'.format,  
    'SEK': '{:<9}'.format,  
    'BRANCH': '{:<7}'.format,
    'LD': '{:<5}'.format,  
    'SD': '{:<5}'.format,
    'GSC': '{:<12}'.format,  
    'TYPE': '{:<13}'.format,
    'N10': '{:<7}'.format,  
    'N': '{:<7}'.format,
}
```

```python
ephem_file = " CONS NAME             PS RATING RA2000   DE2000  M0           PER          MAX      PRIM     SEK      BRANCH LD   SD   GSC         TYPE         N10    N"
for i in range(0, len(ephem_df)):
    ephem_file += "\n"
    ephem_file += f' {ephem_df.iloc[i].CONS}  {ephem_df.iloc[i].NAME.ljust(16)} {ephem_df.iloc[i].PS.ljust(3)}'
    ephem_file += f'{str(ephem_df.iloc[i].RATING).ljust(6)} {str(ephem_df.iloc[i].RA2000).ljust(8)} {str(ephem_df.iloc[i].DE2000).ljust(7)}'
    ephem_file += f' {str("{:12.6f}".format(ephem_df.iloc[i].M0)).ljust(12)} {str("{:11.9f}".format(ephem_df.iloc[i].PER)).ljust(12)} {str(ephem_df.iloc[i].MAX).ljust(8)}'
    ephem_file += f' {str(ephem_df.iloc[i].MAX).ljust(8)}' #PRIM
    ephem_file += f' {str(ephem_df.iloc[i].MAX).ljust(8)}' #SEK
    ephem_file += f' {str(ephem_df.iloc[i].BRANCH).ljust(6)}' #BRANCH
    ephem_file += f' {str(ephem_df.iloc[i].LD).ljust(4)}' #LD
    ephem_file += f' {str(ephem_df.iloc[i].SD).ljust(4)}' #SD
    ephem_file += f' {str(ephem_df.iloc[i].GSC).ljust(11)}' #GSC
    ephem_file += f' {str(ephem_df.iloc[i].TYPE).ljust(12)}' #TYPE
    ephem_file += f' {str(ephem_df.iloc[i].N10).ljust(6)}' 
    ephem_file += f' {str(ephem_df.iloc[i].N).ljust(6)}' 
```

```python
with open('output.dat', 'w') as file:
    # Write the string to the file
    file.write(ephem_file)
```
